### 12) WAP to find whether the given number is Prime or not.

```python
n = int(input("Enter the number:"))
count=0
for i in range(1,n+1) :
    if(n%i==0) :
        count=count+1
if(count==2) :
    print("Prime Number")
else :
    print("Not Prime Number")
```

### 13) WAP to print sum of digits of given number.

```python
n = int(input("Enter the number:"))
sum=0
while(n!=0) :
    rem=n%10
    sum = sum + rem
    n=n//10
print(f"Sum={sum}")
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
n = int(input("Enter the number:"))
rev=0;num=n
while(num!=0) :
    rem=num%10
    rev=rev*10 + rem
    num=num//10
if(rev==n) :
    print("Palindrome Number")
else :
    print("Not Palindrome Number")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
import math
n = int(input("Enter the number:"))
sum=0;num=n;digit=0;x=n
while(x!=0) :
    x=x//10
    digit+=1
while(num!=0) :
    rem=num%10
    sum = sum + math.pow(rem,digit)
    num=num//10
if(sum==n) :
    print("Armstrong Number")
else :
    print("Not Armstrong Number")
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
n = int(input("Enter the number:"))
for i in range(1,n+1) :
    sum=0
    for j in range(1,i) :
        if(i%j==0) :
            sum = sum + j
    if(sum==i) :
        print(i,end=" ")
```
