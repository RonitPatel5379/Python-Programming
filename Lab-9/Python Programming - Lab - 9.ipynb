### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
n = int(input("Enter the number:"))
def isPrime(n) :
    count=0
    for i in range(1,n+1) :
        if(n%i==0) :
            count+=1
    if(count==2) :
        return 1
    else :
        return 0
print(isPrime(n))
```

### 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
start = int(input("Enter the number-1:"))
end = int(input("Enter the number-2:"))
def get_primes(start, end):
    prime_list = []
    for num in range(start, end + 1):
        if num > 1:
            for i in range(2, int(num**0.5) + 1):
                if (num % i) == 0:
                    break
            else:
                prime_list.append(num)
    return prime_list
print(get_primes(start,end))
```

### 08) Write a recursive function to find nth term of Fibonacci Series.

```python
n = int(input("Enter the number:"))
def fibbo(n) :
    if n==0 :
        return 0
    elif n==1 :
        return 1
    else :
        return fibbo(n-1) + fibbo(n-2)
print(fibbo(n-1))
```

### 10) Write a function to get the sum of the scores ending with zero.

### 12) Write a function that returns the number of uppercase and lowercase letters in the given string. 
#### example : Input : s1 = AbcDEfgh ,Ouptput : no_upper = 3, no_lower = 5

```python
s1 = input("Enter the string:")
def noOfUpperLower(s1) :
    upper = 0;lower = 0
    for i in s1 :
        if i.isupper() :
            upper +=1
        else :
            lower +=1
    return upper,lower
upper,lower = noOfUpperLower(s1)
print(upper)
print(lower)
```

### 15) For the given list of names of students, convert the first letter of all the names into uppercase. use map().

### 16) Write udfs to call the functions with following types of arguments:
1. Positional Arguments
2. Keyword Arguments
3. Default Arguments
4. Variable Legngth Positional(*args) & variable length Keyword Arguments (**kwargs)
5. Keyword-Only & Positional Only Arguments

```python
# 1. Positional Arguments
def add(a, b):
    print("Sum=", a + b)
add(10, 20)
# 2. Keyword Arguments
def add(a,b) :
    print("Sum=", a + b)
add(b=20,a=10)
# 3. Default Arguments
def greet(name, msg="Good Morning"):
    print(msg, name,end=",")
greet("Amit")
greet("Ravi", "Good Evening")
# 4. Variable Length Positional Arguments (*args) & variable length Keyword Arguments (**kwargs)
def total(*numbers):
    s = 0
    for i in numbers:
        s += i
    print("\nTotal:", s)
total(10, 20, 30, 40)
def show_details(**info) :
    for key, value in info.items() :
        print(key, ":", value,end=",")
show_details(name="Ajay", age=22, city="Rajkot")
# 5. Keyword-Only Arguments & Positional Only Arguments
def divide(a, b, /):
    print("\nResult:", a / b)
divide(10, 2)
def login(*, username, password):
    print("Username:", username,end=",")
    print("Password:", password)
login(username="admin", password="1234")
```
