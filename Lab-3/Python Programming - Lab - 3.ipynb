### 01) WAP to check whether the given number is positive or negative.

```python
num = int(input("Enter a number:"))
if(num>0) :
    print("num is positive.")
elif(num<0) :
    print("num is negative.")
else :
    print("num is zero.")
```

### 03) WAP to find out largest number from given two numbers using simple if and ternary operator.

```python
a = int(input("Enter first number:"))
b = int(input("Enter second number:"))
if(a>b) :
    print(f"{a} is largest.")
else :
    print(f"{b} is largest.")
ans = a if a>b else b
print(f"{ans} is largest.")
```

### 05) WAP to check whether the given year is leap year or not.
#### [If a year can be divisible by 4 but not divisible by 100 then it is leap year but if it is divisible by 400 then it is leap year]

```python
year = int(input("Enter the year:"))
if(year%4==0 and year%100!=0 or year%400==0) :
    print(f"{year} is leap year.")
else :
    print(f"{year} is not leap year.")
```

### 06) WAP in python to display the name of the day according to the number given by the user.

```python
day = int(input("Enter the day:"))
match day :
    case 1 :
        print("Sunday")
    case 2 :
        print("Monday")
    case 3 :
        print("Tuesday")
    case 4 :
        print("Wednesday")
    case 5 :
        print("Thursday")
    case 6 :
        print("Friday")
    case 7 :
        print("Saturday")
    case _ :
        print("Invalid number")
```

```python
day = int(input("Enter the day:"))
if(day<0) :
    print("Invalid input number.")
else :
    day = day%7
    match day :
        case 1 :
            print("Sunday")
        case 2 :
            print("Monday")
        case 3 :
            print("Tuesday")
        case 4 :
            print("Wednesday")
        case 5 :
            print("Thursday")
        case 6 :
            print("Friday")
        case 0 :
            print("Saturday")
```

### 07) WAP to implement simple calculator which performs (add,sub,mul,div) of two no. based on user input.

```python
a = int(input("Enter first number:"))
b = int(input("Enter second number:"))
operation = input("Enter the operation(+,-,*,/):")
if(operation=='+') :
    print(f"Sum of {a} and {b} is {a+b}.")
elif(operation=='-') :
    print(f"Subtraction of {a} and {b} is {a-b}.")
elif(operation=='*') :
    print(f"Multiplication of {a} and {b} is {a*b}.")
elif(operation=='/') :
    if(b!=0) :
        print(f"Division of {a} of {b} is {a/b}.")
    else :
        print("Division is not possible.")
else :
    print("Invalid number or operation.")
```

### 08) WAP to read marks of five subjects. Calculate percentage and print class accordingly. 
#### Fail below 35 
#### Pass Class between 35 to 45 
#### Second Class
#### between 45 to 60
#### First Class between 60 to 70
#### Distinction if more than 70

```python
a = int(input("Enter the marks of sub-1:"))
b = int(input("Enter the marks of sub-2:"))
c = int(input("Enter the marks of sub-3:"))
d = int(input("Enter the marks of sub-4:"))
e = int(input("Enter the marks of sub-5:"))
percentage = (a+b+c+d+e)/5
if(percentage>70) :
    print("Distinction")
elif(percentage>60) :
    print("First Class")
elif(percentage>45) :
    print("Second Class")
elif(percentage>35) :
    print("Pass Class")
else :
    print("Fail")
```

### 10) WAP to calculate electricity bill based on following criteria. Which takes the unit from the user.
#### a. First 1 to 50 units – Rs. 2.60/unit
#### b. Next 50 to 100 units – Rs. 3.25/unit
#### c. Next 100 to 200 units – Rs. 5.26/unit
#### d. above 200 units – Rs. 8.45/unit

```python
a = float(input("Enter the used units:"))
bill = 0
if(a>=1 and a<=50) :
    bill = a*2.60
if(a>50 and a<=100) :
    bill = 50*2.60 + (a-50)*3.25
if(a>100 and a<=200) :
    bill = 50*2.60 + 50*3.25 + (a-100)*5.26
if(a>200) :
    bill = 50*2.60 + 50*3.25 + 100*5.26 + (a-200)*8.45
print(f"{bill}")
```
