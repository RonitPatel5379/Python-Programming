### 04) WAP to interchange first and last elements in a list.

```python
n = int(input("Enter the size of the list:"))
l1 = list()
for i in range(n):
    element = int(input("Enter the elements:"))
    l1.append(element)
if(n>1) :
    temp = l1[0]
    l1[0] = l1[-1]
    l1[-1] = temp
print(l1)
```

### 05) WAP to split the List into two parts and append the first part to the end.

```python
n = int(input("Enter the size of the list:"))
l1 = list()
mid_index = n//2
for i in range(n) :
    element = int(input("Enter the elements:"))
    l1.append(element)
for i in range(mid_index):
    temp = l1.pop(0)
    l1.append(temp)
print(l1)
```

### 12) WAP to count occurrences of an element in a list.

```python
n = int(input("Enter number of elements: "))
l1 = list()
for i in range(n):
    value = int(input(f"Enter element {i+1}: "))
    l1.append(value)
element = int(input("Enter the element to count: "))
count = l1.count(element)
print(f"The element {element} occurs {count} times in the list.")
```

### 14) WAP to extract elements with frequency greater than K.

```python
n = int(input("Enter number of elements: "))
l1 = list()
for i in range(n):
    value = int(input(f"Enter element {i+1}: "))
    l1.append(value)
k = int(input("Enter the value of K: "))
l2 = list()
for item in set(l1):
    if l1.count(item) > k:
        l2.append(item)
print("Elements with frequency greater than K:",l2)
```

### 16) WAP to create a new list (fruit whose name starts with 'b') from the list of fruits given by user.

```python
n = int(input("Enter number of fruits: "))
a_fruits = list()
for i in range(n):
    fruit = input(f"Enter fruit {i+1}: ")
    a_fruits.append(fruit)
b_fruits = list()
for fruit in a_fruits:
    if fruit.lower().startswith('b'):
        b_fruits.append(fruit)
print("Fruits starting with 'b':", b_fruits)
```

### 17) WAP to create a list of common elements from given two lists.

```python
n1 = int(input("Enter number of elements in first list: "))
l1 = list()
for i in range(n1):
    l1.append(int(input(f"Enter element {i+1} of first list: ")))
n2 = int(input("Enter number of elements in second list: "))
l2 = list()
for i in range(n2):
    l2.append(int(input(f"Enter element {i+1} of second list: ")))
common = list()
for item in l1:
    if item in l2 and item not in common:
        common.append(item)
print("Common elements:", common)
```
