### 05) WAP to Merge two dictionaries. If a key exists in both dictionaries, add their values.

**Sample Input:**
```
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}

```
**Sample Output:**
```
{1: 10, 2: 60, 3: 80, 4: 60}
```

````python
d1 = {}
d2 = {}

n1 = int(input("Enter the size of the dictionary-1: "))
for i in range(n1):
    k = int(input("Enter the key:"))
    v = int(input("Enter the value:"))
    d1[k] = v
n2 = int(input("Enter the size of the dictionary-2: "))
for i in range(n2):
    k = int(input("Enter the key:"))
    v = int(input("Enter the value:"))
    d2[k] = v

for k in d2:
    d1[k] = d1.get(k,0) + d2[k]

print(d1)
````

### 07) WAP to find the keys having maximum and minimum values in a dictionary.

**Sample Input:**
```
{'A': 50, 'B': 90, 'C': 30}
```

```

Min Value Key: C

```

````python
d1 = {}
n = int(input("Enter the size of the dictionary-1: "))
for i in range(n):
    k = input("Enter the key:")
    v = int(input("Enter the value:"))
    d1[k] = v

max_key = max(d1,key=d1.get)
min_key = min(d1,key=d1.get)

print(f"Max Value Key: {max_key}")
print(f"Min Value Key: {min_key}")
````

### 11) WAP to find the student with the highest total marks, given student names and their marks.

**Sample Input:**
```
{'Amit': [70, 80, 90], 'Riya': [85, 75, 80], 'Neha': [60, 70, 65]}
```
**Sample Output:**
```
Topper: Amit
```

````python
n = int(input("Enter number of students: "))
students = {}

for i in range(n):
    name = input("Enter student name: ")
    marks = list(map(int, input("Enter the marks: ").split()))
    students[name] = marks

topper = ""
max_total = 0

for name, marks in students.items():
    total = sum(marks)
    if total > max_total:
        max_total = total
        topper = name

print(f"Topper: {topper}")
````

### 12) WAP to extract keys whose values are even numbers.

**Sample Input:**
```
{1: 11, 2: 20, 3: 15, 4: 40}
```

```
[2, 4]
```

````python
d1 = {}
n = int(input("Enter the size of the dictionary-1: "))
for i in range(n):
    k = int(input("Enter the key:"))
    v = int(input("Enter the value:"))
    d1[k] = v

even_values = []

for k,v in d1.items():
    if v%2 == 0:
        even_values.append(k)
print(even_values)
````

### 15) WAP to create a dictionary using dictionary comprehension from two lists containing student names and marks. Include only those students in the dictionary whose marks are greater than or equal to 75.

**Sample Input:**
```
students = ['Amit', 'Riya', 'Neha', 'Karan']
marks = [78, 72, 85, 60]
```
**Sample Output:**
```
{'Amit': 78, 'Neha': 85}
```

````python
n = int(input("Enter number of students: "))
students = []
marks = []
for i in range(n):
    students.append(input("Enter students name: "))
    marks.append(int(input("Enter marks: ")))
result = {students[i]: marks[i] for i in range(n) if marks[i] >= 75}
print(result)
````
