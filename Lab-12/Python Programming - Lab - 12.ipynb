### 01) WAP to handle following exceptions:
1. ZeroDivisionError
2. ValueError
3. TypeError
#### Note: handle them using separate except blocks and also using single except block too.

```python
a = 10
b = 0

try:
    result = a / b
    print(f"Result: {result}")
except ZeroDivisionError:
    print("Error: You cannot divide by zero!")
try:
    n = int("Hello World")
except ValueError:
    print("Error: That string cannot be converted to a number.")
try:
    total = "Total: " + 100
except TypeError:
    print("Error: You cannot mix strings and numbers in this operation.")
except :
    print("Error occured while trying to handle the error...")
```

### 05) WAP to demonstrate else  and finally block.

```python
balance = 5000
debit = int(input("Enter amount to debit:"))
try:
    if(debit>balance or balance<0) :
        raise ValueError("Debit is greater than balance")
except ValueError as err:
    print(err)
else:
    balance = balance - debit
    print(f"Remaining balance is = {balance}")
finally:
    print("Transaction Completed...")
```

### 07) WAP to accept item prices and calculate total. Raise ValueError if price is negative, otherwise print the total bill amount.

```python
def calculate_total() :
    try:
        n = int(input("Enter no. of items:"))
        total=0

        for i in range(n):
            price = int(input(f"Enter price of item {i+1}:"))

            if(price<0) :
                raise ValueError("Price can't be negative!")
            total+=price
        print("Total Bill amount:",total)
    except ValueError as e:
         print(e)
calculate_total()
```

### 08) Create a short program that prompts the user for a list of grades separated by commas. 
#### Split the string into individual grades and use a list comprehension to convert each string to an integer. 
#### You should use a try statement to inform the user when the values they entered cannot be converted.

```python
try:
    grade = input("Enter the grades with coma seprate").split(',')

    int_grade = [int(x) for x in grade if int(x)>0 and int(x)<=100]

    print(int_grade)
    
except Exception as err:
    print(err)
finally:
    print("Execution completed")
```

### 09) Accept 5 subject marks (0-100). Use assert for validation.

### 10) WAP that gets password from user. Password must have at least 8 characters and one digit. 
### Raise WeakPasswordError if invalid, print the messsage "Correct" otherwise.

```python
class WeakPasswordError(Exception):
    pass
password = input("Enter your password: ")
try:
    if len(password) < 8:
        raise WeakPasswordError("Password must be at least 8 characters long.")
    if not any(char.isdigit() for char in password):
        raise WeakPasswordError("Password must contain at least one digit.")
    print("Correct")
except WeakPasswordError as e:
    print("WeakPasswordError:", e)
```
