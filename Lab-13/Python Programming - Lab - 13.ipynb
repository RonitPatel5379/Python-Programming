### 08) Subtract a week (7 days) from a given date in Python.

### 09) WAP to Calculate number of days between two given dates.

```python
from datetime import date
def calculate_days(date1, date2):
    delta = abs(date2 - date1)
    return delta.days
d1 = date(2026, 3, 1)
d2 = date(2025, 3, 3)
result = calculate_days(d1, d2)
print(f"The number of days between {d1} and {d2} is: {result} days")
```

### 10) WAP to Find the day of the week of a given date.(i.e. wether it is sunday/monday/tuesday/etc.)

```python
import datetime
given_date = datetime.date(2026, 3, 3)
day_name = given_date.strftime("%A")
print(f"The day of the week is: {day_name}")
```
