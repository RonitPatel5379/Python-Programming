### 02) WAP to find Maximum and Minimum K elements in a given tuple.
**Sample Input:**
```
T = (7, 2, 9, 4, 1, 5)
K = 2

```
**Sample Output:**
```
Minimum 2 elements: (1, 2)
Maximum 2 elements: (9, 7)
```

````python
n = int(input("Enter the size of tuple:"))
K = int(input("Enter the value of K:"))
t1 = ()
for i in range(n) :
    t1 += tuple(map(int,input("Enter the elements:").split()))
s = tuple(sorted(t1))
min_k = tuple(s[:K])
max_k = tuple(s[-K:][::-1])
print(f"Minimum {K} elements:{min_k}")
print(f"Maximum {K} elements:{max_k}")
````

### 03) WAP to find tuples which have all elements divisible by K from a list of tuples.
**Sample Input:**
```
tuple_list = [(2, 4, 6), (3, 6, 9), (4, 8, 12), (5, 10, 15)]
K = 2

```
**Sample Output:**
```
Tuples divisible by 2 : [(2, 4, 6), (4, 8, 12)]
```

````python
n = int(input("Enter number of tuples: "))
tuple_list = []
for i in range(n):
    t = tuple(map(int, input("Enter elements of tuple separated by space: ").split()))
    tuple_list.append(t)
K = int(input("Enter value of K: "))
result = []
for t in tuple_list:
    if all(x % K == 0 for x in t):
        result.append(t)
print("Tuples with all elements divisible by", K, "are:", result)
````

### 14) WAP to create (index, value) pairs for prime numbers.

**Sample Input:**
```
T = (4, 7, 9, 11, 15)
```
**Sample Output:**
```
((1, 7), (3, 11))
```

````python
n = int(input("Enter the size of tuple-1:"))
t1 = tuple()
for i in range(n) :
    t1 += tuple(map(int, input("Enter the elements: ").split()))
result = []
for index, value in enumerate(t1):
    if value > 1:
        is_prime = True
        for i in range(2, int(value ** 0.5) + 1):
            if value % i == 0:
                is_prime = False
                break
        if is_prime:
            result.append((index, value))
print(tuple(result))
````

### 16) WAP to compute column-wise sum from a tuple of lists.

**Sample Input:**
```
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])
```

**Sample Output:**
```
12 15 18

```

````python
n = int(input("Enter number of lists: "))
t = []
for i in range(n):
    lst = list(map(int, input("Enter list elements: ").split()))
    t.append(lst)
t = tuple(t)
cols = len(t[0])
result = []
for j in range(cols):
    col_sum = 0
    for i in range(len(t)):
        col_sum += t[i][j]
    result.append(col_sum)
for val in result:
    print(val, end=" ")
````
